```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
school_data.head()
```

```python
# Read Student Data File and store into Pandas DataFrames
student_data = pd.read_csv(student_data_to_load)
student_data.head()
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name"])
school_data_complete.head()
```

```python
#Calculate the total number of schools
total_schools = len(school_data_complete['School ID'].unique())
total_schools
```

```python
#Calculate the total number of students
total_students = len(school_data_complete['Student ID'].unique())
total_students
```

```python
#Calculate the total budget
total_budget =school_data['budget'].sum()
total_budget
```

```python
#Calculate the average maths score
average_maths_score =  school_data_complete['maths_score'].mean()
average_maths_score
```

```python
#Calculate the average reading score
average_reading_score =  school_data_complete['reading_score'].mean()
average_reading_score
```

```python
#Calculate the percentage of students with a passing maths score (50 or greater)
pct_passing_maths = len(school_data_complete.loc[school_data_complete['maths_score']>=50])/total_students * 100
pct_passing_maths
```

```python
#Calculate the percentage of students with a passing reading score (50 or greater)
pct_passing_reading = len(school_data_complete.loc[school_data_complete['reading_score']>=50])/total_students * 100
pct_passing_reading
```

```python
#Calculate the percentage of students who passed maths and reading (% Overall Passing)
pct_overall_passing = len(school_data_complete.loc[(school_data_complete['maths_score']>=50) &\
                                                   (school_data_complete['reading_score']>=50)])/total_students * 100
pct_overall_passing
```

```python
# store results in a Dataframe
area_summary = pd.DataFrame({'Total Schools' : [total_schools],
                           'Total Students' : [total_students],
                            'Total Budget' : [total_budget],
                            'Average Maths Score' : [average_maths_score],
                             'Average Reading Score' : [average_reading_score],
                            '% Passing Maths' : [pct_passing_maths],
                            '% Passing Reading' : [pct_passing_reading],
                            '% Overall Passing' : [pct_overall_passing]})
#formatting
area_summary['Total Students'] = area_summary['Total Students'].map("{:,}".format)
area_summary['Total Budget'] = area_summary['Total Budget'].map("${:,.2f}".format)
area_summary
```

```python
school_data_complete['Passing Maths'] = school_data_complete['maths_score'].apply(lambda x: 1 if x >=50 else 0)
school_data_complete['Passing Reading'] = school_data_complete['reading_score'].apply(lambda x: 1 if x >=50 else 0)
```

```python
school_data_complete['Overall Passing'] = np.where((school_data_complete['Passing Maths'] == 1) &\
                                                   (school_data_complete['Passing Reading'] == 1),1,0)
school_data_complete
```

```python
school_gb = school_data_complete.groupby(['school_name'])
school_gb.count().head()
```

```python
ave_scores_school = school_gb[['reading_score','maths_score']].mean()
ave_scores_school
```

```python
total_passing_school = school_gb[['Passing Maths','Passing Reading', 'Overall Passing']].sum()
total_passing_school
```

```python
full_school_data = school_data.merge(ave_scores_school, on ="school_name").merge(total_passing_school, on ="school_name")
full_school_data
```

```python
full_school_data['per_student_budget'] = full_school_data['budget']/full_school_data['size']
full_school_data['% Passing Maths'] = full_school_data['Passing Maths']/full_school_data['size'] * 100
full_school_data['% Passing Reading'] = full_school_data['Passing Reading']/full_school_data['size'] * 100
full_school_data['% Overall Passing'] = full_school_data['Overall Passing']/full_school_data['size'] * 100
```

```python
per_school_summary = full_school_data[['school_name', 'type', 'size', 'budget', 'per_student_budget','maths_score','reading_score','% Passing Maths', '% Passing Reading', '% Overall Passing']]
per_school_summary.columns = ['school_name', 'School Type', 'Total Students', 'Total Budget', 'Per Student Budget','Average Maths Score','Average Reading Score','% Passing Maths', '% Passing Reading', '% Overall Passing']
per_school_summary = per_school_summary.sort_values(by='school_name').set_index('school_name')

#formatting
per_school_summary['Total Students'] = per_school_summary['Total Students'].map("{:,}".format)
per_school_summary['Total Budget'] = per_school_summary['Total Budget'].map("${:,.2f}".format)
per_school_summary['Per Student Budget'] = per_school_summary['Per Student Budget'].map("${:,.2f}".format)
per_school_summary
```

Maths Scores by Year
Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

Create a pandas series for each year. Hint: use a conditional statement.

Group each series by school

Combine the series into a dataframe

Optional: give the displayed data cleaner formatting

```python
ave_maths_score_years = []
for year in school_data_complete['year'].unique():
    temp  = school_data_complete.loc[school_data_complete['year']==year,['school_name','maths_score']]
    temp_gb = temp.groupby(['school_name'])['maths_score'].mean().reset_index(name= 'Year ' + str(year))
    temp_gb.set_index('school_name',inplace=True)
    ave_maths_score_years.append(temp_gb)
```

```python
#create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.
maths_score_by_year  = pd.concat(ave_maths_score_years, axis=1)
maths_score_by_year = maths_score_by_year[['Year 9', 'Year 10','Year 11','Year 12']]
maths_score_by_year
```

Reading score by year
Perform the same operations as above for reading scores

```python
ave_reading_score_years = []
for year in school_data_complete['year'].unique():
    temp  = school_data_complete.loc[school_data_complete['year']==year,['school_name','reading_score']]
    temp_gb = temp.groupby(['school_name'])['reading_score'].mean().reset_index(name= 'Year ' + str(year))
    temp_gb.set_index('school_name',inplace=True)
    ave_reading_score_years.append(temp_gb)

#create a table that lists the average reading score for students of each year level (9, 10, 11, 12) at each school.
reading_scores_by_year  = pd.concat(ave_reading_score_years, axis=1)
reading_scores_by_year = reading_scores_by_year[['Year 9', 'Year 10','Year 11','Year 12']]
reading_scores_by_year
```

Scores by School Spending
Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
Average Maths Score
Average Reading Score
% Passing Maths
% Passing Reading
Overall Passing Rate (Average of the above two)

```python
bins = [0,585,630,645,680]
labels = ["<$585","$585-630","$630-645","645-680"]

per_school_summary['Per Student Budget(int)'] = pd.to_numeric(per_school_summary['Per Student Budget'].str.replace("$", ""))
per_school_summary['Spending Ranges (Per Student)'] = pd.cut(per_school_summary['Per Student Budget(int)'],bins=bins, labels=labels)
del per_school_summary['Per Student Budget(int)']
per_school_summary
```

```python
pd.options.display.float_format = '{:.2f}'.format
spending_summary = per_school_summary.groupby('Spending Ranges (Per Student)')[['Average Maths Score','Average Reading Score','% Passing Maths', '% Passing Reading', '% Overall Passing']].mean()
spending_summary
```

Scores by School Size
Perform the same operations as above, based on school size.

```python
bins = [0,1000,2000,5000]
labels = ["Small (<1000)","Medium (1000-2000)","Large (2000-5000)"]
per_school_summary['Total Students(int)'] = pd.to_numeric(per_school_summary['Total Students'].str.replace(",", ""))
per_school_summary['School Size'] = pd.cut(per_school_summary['Total Students(int)'],bins=bins, labels=labels)
size_summary = per_school_summary.groupby('School Size')[['Average Maths Score','Average Reading Score','% Passing Maths', '% Passing Reading', '% Overall Passing']].mean()
size_summary
```
